```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carica il file .npy
file_path = "C:/Users/admin/Desktop/miccai_2025_ste/final_repo/new_final/masks/train_mask_epoch_26.npy"

data = np.load(file_path)

# Crea la heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data[56], cmap="viridis", annot=False)  # Usa "coolwarm" o "plasma" per variare i colori

num_zeri = np.count_nonzero(data == 0)
print(f"Numero di zeri nella matrice: {num_zeri}")


# Mostra il grafico
plt.title("Heatmap del file .npy")
plt.show()
```
